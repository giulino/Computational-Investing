### 1.4 Calculating and Visualizing Returns

Given a dataset representing the annual closing values of an investment portfolio over a five year period with more volatile values. 
Calculate both the arithmetic and logarithmic returns for each year, the total returns over the entire period, and visualize the cumulative returns to see the difference visually.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

closing_values = np.array([10, 12.5, 8, 13.5, 7.5, 15]) 

# Calculate the arithmetic return
arithmetic_return = (closing_values[1:] - closing_values[:-1]) / closing_values[:-1]
arithmetic_return

# Calculate the log return
log_return = np.log(closing_values[1:] / closing_values[:-1])
log_return

plt.figure(figsize=(10,6))
plt.plot(log_return, label = "Log Return", marker = "o", markersize = "10", markerfacecolor = "green", markeredgecolor = "green")
plt.plot(arithmetic_return, label = "Arithmetic Return", marker = "o", markersize = "10", markerfacecolor = "red", markeredgecolor = "red")
plt.title("Arithemetic vs Logarithmic Return")
plt.xlabel('Year')
plt.ylabel("Return")
plt.legend()
plt.grid(True) #color = "black"
plt.show()
```

### Compare and Discuss

The arithmetic return line shows greater amplitude in its fluctuations compared to the logarithmic return line. The reason behind this is the linearity progression of these returns.  

### 1.4.2 Calculate and Plot Cumulative Returns

```python
cumulative_arithmetic_return = np.cumsum(arithmetic_return)
cumulative_log_return = np.exp(np.cumsum(log_return)) - 1 

plt.figure(figsize=(12,6))
plt.plot(cumulative_arithmetic_return, label = "Cumulative Arithmetic Return", marker = "o")
plt.plot(cumulative_log_return, label = "Cumulative Log Return", marker = "o")
plt.title("Cumulative Returns")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Return")
plt.grid(True)
plt.show()
```

### Compare and Discuss

In this graph, we can clearly notice the differences between these two calculations: the arithmetic return simply add-up without considering the compunding effett whilst the log return offers a clearer rapresentation of the performance over time. This is evident between year 2 and year 3. The magnitude of the loss is greater compared to the arithmetic return. This is due the non-linear progression of the log returns. 

In addition, we can clearly state that log returns take into account the impact of gains and losses on the evolving investment.
