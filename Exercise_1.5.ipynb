### S&P 500 Futures Data Analysis

Downloading daily data of S&P 500 front month futures from Yahoo Finance and perform a arithmetic and log return calculation

```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ticker = "ES=F"
dataset = yf.download(ticker)
sp_500 = pd.DataFrame(dataset)
sp_500
```

```python
sp500_arithmetic_return = (sp_500['Adj Close'] - sp_500['Adj Close'].shift(1)) / sp_500['Adj Close'].shift(1)
sp500_log_return = np.log(sp_500['Adj Close'] / sp_500['Adj Close'].shift(1))

sp500_cumulative_arithmetic_return = np.cumsum(sp500_arithmetic_return)
sp500_cumulative_log_return = np.exp(np.cumsum(sp500_log_return)-1)

# 

plt.figure(figsize=(12,6))
plt.plot(sp500_arithmetic_return, label = "SP500 Arithmetic Return")
plt.plot(sp500_log_return, label = "SP500 Log Return")
plt.title('SP500 Arithmetic vs Log Return')
plt.xlabel('Year')
plt.ylabel('Return')
plt.legend()
plt.grid(True)
plt.show()

# Cumulative Returns

plt.figure(figsize=(12,6))
plt.plot(sp500_cumulative_arithmetic_return, label = "SP500 Arithmetic Return")
plt.plot(sp500_cumulative_log_return, label = "SP500 Log Return")
plt.title('SP500 Arithmetic vs Log Return')
plt.xlabel('Year')
plt.ylabel('Return')
plt.legend()
plt.grid(True)
plt.show()
```

Looking at the graph the growth of the US stock massive over the last 24 years was massive. However, we can have a different view of the market if look at the log returns or the arithemetic return of the SP500. The log return plot represents a compounded return profile, and it lies below the arithmetic return when plotted on the same chart because the arithmetic return simply measures the sum of periodic returns without accounting for compounding.

The arithmetic return is higher because the arithmetic return is not adjusting for the compounding effect and is therefore overstating the cumulative return compared to the log return.
