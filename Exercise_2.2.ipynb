### Downloading S&P500 Futures Prices and Calculating Annualized Volatility

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "ES=F"
dataset = yf.download(ticker)
sp500= pd.DataFrame(dataset)
sp500
```

```python
sp500["log_return"] = np.log(sp500['Adj Close']).diff()
volatility = sp500.log_return.std()

# Annualize the volatility
annualized_volatility = np.round(volatility * np.sqrt(252) * 100, 2)
print(f"Annual Volatility: {annualized_volatility}%")
```
