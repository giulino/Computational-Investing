### Downloading Nasdaq Futures Prices and Calculating Annualized Mean Logarithmic Return

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "NQ=F"
dataset = yf.download(ticker)
nasdaq = pd.DataFrame(dataset)
nasdaq
```

```python
nasdaq['log_return'] = np.log(nasdaq['Adj Close'] / nasdaq['Adj Close'].shift(1))
annualized_log_return = round((nasdaq['log_return'].mean() * 252)*100, 2)
print(f"Annualized Return {annualized_log_return} % ")
```
